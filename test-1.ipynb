```python
import requests
import json

url = 'https://query1.finance.yahoo.com/v8/finance/chart/amzn?period1=1697760000&period2=1697846399&interval=1d&includePrePost=true&events=split'
headers = { 'User-Agent': 'PostmanRuntime/7.33.0', 'Content-Type': 'application/json'}
response = requests.get(url, headers=headers)

print(response.status_code)

json_object = json.loads(response.content)
currency = json_object['chart']['result'][0]['meta']['currency']

print(currency)

json_formatted_str = json.dumps(json_object, indent=2)
print(json_formatted_str)
print(json_object)
```
